```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
rock_mine = pd.read_csv('sonarmine.csv')
```

```python
rock_mine
```

```python
rock_mine.describe()
```

```python
rock_mine['R'].value_counts()
```

```python
val = list(range(61))
```

```python
rock_mine.columns=val
```

```python
rock_mine
```

```python
rock_mine.shape
```

```python
rock_mine.describe()
```

```python
X = rock_mine.drop(columns=60,axis=1)
X.shape
```

```python
y = rock_mine[60]
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.13, random_state=1)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
LogModel = LogisticRegression()
```

```python
LogModel.fit(X_train,y_train)
```

```python
y_pred=LogModel.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```
